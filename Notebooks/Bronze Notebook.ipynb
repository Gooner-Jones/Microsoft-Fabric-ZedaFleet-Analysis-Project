# **Bronze Layer Scripts**


### **Reading Data and Loading Tables**

```python
# Customers
df_customers = spark.read.format("csv")\
              .option("header", "true")\
              .option("inferSchema", "true")\
              .load("Files/Rawdatasets/crm_data/customers.csv")

df_customers.write.format("delta")\
            .mode("overwrite")\
            .saveAsTable("customers")
```
